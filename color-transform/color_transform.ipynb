## Chuyển đổi ảnh màu sang ảnh xám

$$ gray = \alpha . blue  + \beta . green + \gamma . red$$

Trong đó:

$\alpha = 0.114$ <br>
$\beta = 0.587$ <br>
$\gamma = 0.299$ <br>


```python
def generatingWeight(alpha, beta, gamma):
    return [gamma, beta, alpha]

def color2gray(image):
    return np.uint8(np.dot(img[..., :3], generatingWeight(0.299, 0.587, 0.114)))

original_image = read_image('./data/flower.jpg')
grayscale_image = color2gray(original_image)

# visualize
plt.figure(figsize = (20,10))
plt.subplot(1,2,1), plt.imshow(cv.cvtColor(original_image, cv.COLOR_BGR2RGB))
plt.title('Input image')
plt.subplot(1,2,2), plt.imshow(grayscale_image, 'gray')
plt.title('Grayscale image')
print(original_image.shape)
print(grayscale_image.shape)
```

## Chuyển ảnh xám sang ảnh màu

```python
def gray2color(image):
    return cv.cvtColor(image, cv.COLOR_GRAY2RGB)

original_image = read_image('./data/girl.jpg')
color_image = gray2color(original_image)

# visualize
plt.figure(figsize = (20,10))
plt.subplot(1,2,1), plt.imshow(original_image, 'gray')
plt.title('Input image')
plt.subplot(1,2,2), plt.imshow(color_image)
plt.title('Color image')
print(original_image.shape)
print(color_image.shape)
```
